```Finding loss and error in prediction using MSE - Mean Squared Error```

$MSE = (\hat{y} - {y})^2$

```for best prediction MSE must be as close to zero as possible```

````python
import numpy as np

loss = lambda w,y: ((w * 80 + 11.99) - y) ** 2
y = 81
ws = np.arange(-1,3, .1)
losses = loss(ws,y)

plt.scatter(ws,losses)
plt.plot(1,loss(1,y),'ro')
````

```The Gradient tells how quickly the loss changes and this is how we can find out the perfect weight for minimum MSE```

````python
gradient = lambda w,y: ((w * 80 + 11.99) - y) * 2
title = "Higher the Gradient, Greater the change of MSE with small change in weight"
gradients = gradient(ws,y)

plt.scatter(ws,gradients)
plt.plot(1,gradient(1,y),'ro')
plt.title(title)

# Display the plot
plt.show()
````

````python
#how to update weights and biases to get to minimum point!
#to adjust for the change of gradient we use learning rate!

ws = np.arange(-.5, 1.5, .1)
losses = loss(ws,y)

plt.scatter(ws,losses)

plt.plot(1,loss(1,y),'ro')
lr = 5e-5

new_weight = 1 - lr * gradient(1,y) * 80
plt.plot(new_weight, loss(new_weight,y),'go')
````

````python
PREDICTORS = ["tmax" , "tmin", "rain"]
TARGET = "tmax_tomorrow"

np.random.seed(0)
split_data = np.split(data, [int(.7 * len(data)),int(.85 * len(data))])
(train_x, train_y), (valid_x, valid_y), (test_x, test_y) = [[d[PREDICTORS].to_numpy(),d[[TARGET]].to_numpy()] for d in split_data]
````

````python
import math

def init_params(predictors):
    np.random.seed(0)
    weights = np.random.rand(predictors,1) -.5
    biases = np.ones((1,1))
    return [weights,biases]
````

````python
init_params(3)
````

$w_{1} * t_{max} + w_{2} * t_{min} + w_{3} * rain + b$
            ```formula for operation below```

````python
def forward(params,x):
    weights,biases = params
    prediction = x @ weights + biases
    return prediction
````

````python
def mse(actual, predicted):
    return np.mean((actual - predicted) ** 2)

def mse_grad(actual, predicted):
    return predicted - actual
 #easier to change if not *2
````

````python
def backward(params, x , lr, grad):
    w_grad = (x.T / x.shape[0]) @ grad
    b_grad = np.mean(grad, axis=0)
    
    params[0] -= w_grad * lr
    params[1] -= b_grad * lr
    
    return params
````

````python
#training loop
lr = 1e-4
epochs = 5000

params = init_params(train_x.shape[1])

for i in range(epochs):
    predictions = forward(params, train_x)
    grad = mse_grad(train_y, predictions)
    
    params = backward(params, train_x, lr, grad)
    
    if i%100== 0:
        predictions = forward(params, valid_x)
        valid_loss = mse(valid_y,predictions)
        
        print(f"Epoch {i} loss: {valid_loss}")
````
